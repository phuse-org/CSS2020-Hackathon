```python
#Importing necessary modules
import podr_connections
import pandas as pd
import os

#Setting the working directory
os.chdir(r'C:\Users\glwci\Desktop\Workspace\09.PhUSE\CSS 2020\Hackathon')

#Printing the current working directory
os.getcwd()
```

```python
#Establishing the connection to the data server 
podr = podr_connections.podr_connection('phuse_4kyihhju8p')
```

```python
#Reading the ADAE dataset from the data server  from the respective library
adae = podr.read('ADAE', libname='cdisc_pilot_adam')

#Checking the first few rows of adae dataframe 
adae.head()
```

```python
#Just selecting the relevant columns from the adae dataframe
df_ae = adae.loc[:,['AEDECOD','TRTA','USUBJID']]

#Findning the unique counts of subject ids by group of preferred term and treatment
df_ae_n  = df_ae.groupby(['AEDECOD','TRTA'])['USUBJID'].nunique()

#Checking the first 5 rows of the resultant summary of the previous step
df_ae_n.head()
```

```python
#Reading the ADSL data from the data server to find the total count of subjects participating in each treatment arm
adsl = podr.read('ADSL', libname='cdisc_pilot_adam')

#Top 5 rows of adsl dataframe
adsl.head()
```

```python
#Finding the total no. of subjects by treatment arms
trt_sample = adsl.groupby('TRT01P').size()

#Total no. of subjects in the 'Placebo' arm
N_placebo = trt_sample.loc['Placebo']

#Total no. of subjects in the 'Xanomeline High Dose' arm
N_xhd  = trt_sample.loc['Xanomeline High Dose']

#Total no. of subjects in the 'Xanomeline Low Dose' arm
N_xld = trt_sample.loc['Xanomeline Low Dose']

#Printing the sample sizes
print('Placebo:', N_placebo)
print('Xanomeline High Dose:', N_xhd)
print('Xanomeline Low Dose', N_xld)
```

```python
#Reshaping the adae summary dataframe to see the treatment in different columns 
ae_trt  = df_ae_n.reset_index().pivot(index='AEDECOD',columns='TRTA', values='USUBJID').fillna(0)

#Checking the head
ae_trt.head()
```

```python
#Calculating the percentages of aes in each treatment arm
ae_trt['placebo_pct'] = round(ae_trt['Placebo']*100/N_placebo, 2)
ae_trt['xhd_pct'] = round(ae_trt['Xanomeline High Dose']*100/N_xhd, 2)
ae_trt['xld_pct'] = round(ae_trt['Xanomeline Low Dose']*100/N_xld, 2)

#Head
ae_trt.head()
```

```python
#Setting the datatype of columns with just counts to int, in order to remove the decimals
ae_trt_str = ae_trt.astype({
            'Placebo': int,
            'Xanomeline High Dose' : int,
            'Xanomeline Low Dose': int,
            'placebo_pct': float,
            'xhd_pct': float,
            'xld_pct': float
})

#Head
ae_trt.head()
```

```python
#Formating the previous dataframe by concatenating teh columns with count and percentages in required format

ae_trt_str['Placebo'] = ae_trt_str['Placebo'].map(str) + ae_trt_str['placebo_pct'].map('({})'.format)
ae_trt_str['Xanomeline High Dose'] = ae_trt_str['Xanomeline High Dose'].map(str) + ae_trt_str['xhd_pct'].map('({})'.format)
ae_trt_str['Xanomeline Low Dose'] = ae_trt_str['Xanomeline Low Dose'].map(str) + ae_trt_str['xld_pct'].map('({})'.format)

"""
Other ways to concatenate:
ae_trt_str['Placebo'] = ae_trt_str['Placebo'].map(str) + '(' + ae_trt_str['placebo_pct'].map(str) + ')'
ae_trt_str.apply(lambda x:'%s(%s)' % (x['Placebo'],x['placebo_pct']),axis=1)
"""
#Head
ae_trt_str.head()
```

```python
#Getting only the columns of interest
ae_trt_df = ae_trt_str.loc[:,['Xanomeline High Dose','Xanomeline Low Dose','Placebo']].reset_index()

#Formating the column names
ae_trt_df.columns = ['Preferred Term', 
                     'Xanomeline High Dose <br>(N={}) <br>n(%)'.format(N_xhd), 
                     'Xanomeline Low Dose <br>(N={}) <br>n(%)'.format(N_xld), 
                     'Placebo <br>(N={}) <br>n(%)'.format(N_placebo)]
ae_trt_df.head()
```
